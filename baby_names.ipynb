```python
import pandas as pd

filepath = '/home/jake_ockerby/Python/names/yob1880.txt'
data_1880 = pd.read_csv(filepath, names=['name','sex','births'])
data_1880
```

```python
years = range(1880, 2019)
pieces = []

for year in years:
    path = '/home/jake_ockerby/Python/names/yob%d.txt' % year
    frame = pd.read_csv(path, names=['name','sex','births'])
    frame['year'] = year
    pieces.append(frame) # reading all files from 1880-2018 into dataframes, then storing
                         # the dataframes in a list
        
names = pd.concat(pieces, ignore_index=True)# concatenates all dataframes in the list into one dataframe
```

```python
total_births = names.pivot_table('births', index='year', columns='sex', aggfunc=sum)
total_births.tail()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14,6))
plt.title('Total Births by Sex and Year')
plt.ylabel('Total Births')
sns.lineplot(data=total_births)
```

```python
def add_perc(group):
    births = group.births.astype(float)
    group['percentage'] = (births/births.sum())*100
    return group

names = names.groupby(['year','sex']).apply(add_perc)
names # creating a column for birth names as a percentage of total births by year and sex
```

```python
import numpy as np
np.allclose(names.groupby(['year','sex']).percentage.sum(), 100) # checking if all percentages sum to make 100
```

```python
def gettop1000(group):
    return group.sort_values(by='births', ascending=False)[:1000]

grouped = names.groupby(['year','sex'])
top1000 = grouped.apply(gettop1000)
top1000 # creating a dataframe for top 1000 most popular names by year and sex
        # shortens the original dataframe from almost 2 million rows to approx. 280,000
```

```python
top1000.index = top1000.index.droplevel()
total_births = top1000.pivot_table('births', index='year', columns='name', aggfunc=sum)
total_births
```

```python
subset = total_births[['Jake','Jack','Janet','Josephine']]
plt.figure(figsize=(14,6))
plt.title('Total Births by Name and Year')
plt.ylabel('Total Births')
sns.lineplot(data=subset) # plot to summarize how specific names change in popularity over time
```

```python
top1000.index = top1000.index.droplevel()
name_diversity = top1000.pivot_table('percentage', index='year', columns='sex', aggfunc=sum)
name_diversity
```

```python
plt.figure(figsize=(14,6))
plt.title('Percentage of Births Represented in Top 1000 Names by Sex')
plt.xlim(1880,2020)
plt.ylim(0,120)
plt.ylabel('Percentage of Names in Top 1000')
sns.lineplot(data=name_diversity) # plot showing that first names have become much more diverse since 1880,
                                  # especially female names
```
